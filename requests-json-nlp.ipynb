2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Open JSON file
with open('lyrics.json', 'r') as openfile:
    # Reading from json file
    json_object = json.load(openfile)

lyrics = json_object['lyrics']

# Print the lyrics of the song
print(lyrics) 

# Run SpacyTextBlob
nlp = spacy.load('en_core_web_sm')
nlp.add_pipe('spacytextblob')
doc = nlp(lyrics)
polarity = doc._.blob.polarity

# Print polarity score
print("Polarity is: ", polarity)

# Polarity Answer: I believe that the lyrics have a more positive connotation. Using the NLP Sentiment from SpacyTextBlob, I got a score of 0.04505208333333333.
# This score makes it more positive so that is my answer for the lyrics.
```

3. Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

```python
import requests
import json

# Music function
def music(artist, song, filename):
    lyrics = json.loads(requests.get('https://api.lyrics.ovh/v1/' + artist + "/" + song).text)

    # Output dictionary object to JSON file
    with open(filename + '.json', 'w') as outfile:
        json.dump(lyrics, outfile)
    
# Four Songs     
music("Good Kid", "From the Start", "From the Start Lyrics") # Output file name: From the Start Lyrics.json
music("Good Kid", "No Time to Explain", "No Time to Explain Lyrics") # Output file name: No Time to Explain Lyrics.json
music("Forrest Frank", "UP!", "UP! Lyrics") # Output file name: UP! Lyrics.json
music("Andy Mineo", "Coming In Hot", "Coming In Hot Lyrics") # Output file name: Coming In Hot Lyrics.json
```

4. Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score.  Use this function to print the polarity scores (with the name of the song) of the three files you created in question 3.  Does the reported polarity match your understanding of the song's lyrics?
```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

def song(filename):
    # Open JSON file
    with open(filename + '.json', 'r') as openfile:
        # Reading from json file
        json_object = json.load(openfile)

    lyrics = json_object['lyrics']

    # Run SpacyTextBlob
    nlp = spacy.load('en_core_web_sm')
    nlp.add_pipe('spacytextblob')
    doc = nlp(lyrics)
    polarity = doc._.blob.polarity

    print(f"Polarity Score for {filename} is {polarity}")

song("From the Start Lyrics")
song("No Time to Explain Lyrics")
song("UP! Lyrics")
song("Coming In Hot Lyrics")
```
